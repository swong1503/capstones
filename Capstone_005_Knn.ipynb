Use the unknown features or unknown columns name to classify target class

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("GnBu_d")
sns.set_style('whitegrid')
%matplotlib inline
```

```python
#read the data
df = pd.read_csv('KNN_Project_Data')
```

```python
#info of the data
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#Explore the data
#pairplot of df with hue of Target Class
sns.pairplot(data=df, hue='TARGET CLASS', palette='bwr')
```

```python
#Standardize the variables
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS', axis=1))
```

```python
scaled_features = scaler.transform(df.drop('TARGET CLASS', axis=1))
```

```python
#Conver scaled features to a dataframe
df_feat = pd.DataFrame(scaled_features, columns=df.columns[:-1])
```

```python
df_feat.head()
```

```python
#Split data
from sklearn.model_selection import train_test_split
```

```python
X = df_feat
y = df['TARGET CLASS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#Train using Knn
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Create a model using k = 1
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
#Prediction
pred = knn.predict(X_test)
```

```python
#Evaluation
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```

```python
#Choose an optimal K value
#Create a loop 
error_rate = []

for i in range (1,60):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
#plot error rate vs. k values
plt.figure(figsize=(10,6))
plt.plot(range(1,60), error_rate, color='blue', linestyle='dashed',
         marker='o', markerfacecolor='red', markersize=10)
plt.title('Error Rate vs K Value')
plt.xlabel('K Value')
plt.ylabel('Error Rate')
```

```python
min(error_rate)
```

```python
#When K value is 31, 37, 39, 53, or 54, error = 0.16
```

```python
#Use the model again using k value that gives lowest error rate
knn = KNeighborsClassifier(n_neighbors=37)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```

```python
#This increased precision by 12% when k increased from 1 to 37.
```
