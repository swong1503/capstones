```python
#create K Means model with 2 clusters
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2) #We expect two groups private vs public
```

```python
#fit the model
#fit in all the data expect private labels.
kmeans.fit(df.drop('Private', axis=1))
```

```python
#Check the centers
kmeans.cluster_centers_
```

```python
#Evaluation
#There is no perfect way to evaluate clustering if we don't have the labels.
#since we have it in this one, we take advantage to evaluate the clusters.

#create a new column for df called cluster, which 1 for Private and 0 for pub.
def converter(cluster): #Create a function call converter
    if private == 'Yes': #from the data if it is private it returns 1
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter) #apply the function to df.clus.
```

```python
#Confusion matrix and report
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(df['Cluster'], kmeans.labels_))
print('\n')
print(classification_report(df['Cluster'], kmeans.labels_))
```
